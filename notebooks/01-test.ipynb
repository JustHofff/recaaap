# Initial TrOCR Test

```python
# Import libraries
from transformers import TrOCRProcessor, VisionEncoderDecoderModel
from PIL import Image, ImageOps
import requests
from io import BytesIO
from datasets import load_dataset
```

## Loading TrOCR processor and EncoderDecoderModel

```python
# The processor normalizes the input images
# The model encodes the image, then decodes it into a predicted text
processor = TrOCRProcessor.from_pretrained("microsoft/trocr-small-handwritten")
model = VisionEncoderDecoderModel.from_pretrained("microsoft/trocr-small-handwritten")
print('Loading processor and model complete')
```

## Loading the image

```python
# If you have a local image: image = Image.open("your_image.png").convert("RGB")
#dataset = load_dataset("nielsr/IAM-data", split="train[:1]")
#image = dataset[0]["image"].convert("RGB")

image = Image.open("/Users/justinhoffman/IMG_6847.jpg").convert("RGB")
image = ImageOps.grayscale(image)
image = ImageOps.autocontrast(image)

image = image.convert("RGB")
print('Loading image complete')
```

## Encoding and Decoding 

```python
# Preprocess the image
pixel_values = processor(images=image, return_tensors="pt").pixel_values
print('Preprocessing image complete')

# Generate text
generated_ids = model.generate(pixel_values)
print('Generating text complete')

# Decode tokens back to a string
generated_text = processor.batch_decode(generated_ids, skip_special_tokens=True)[0]

print("Predicted text:", generated_text)
```

## Testing
Inserted a picture of the word: 'HELP'
Model returned the predicted text: 'if # #'

After adding extra preprocessing:
- converting image to grayscale
- maximizing the image contrast
- converting back to RGB,
Model returned the predicted text: '# Help'

## Findings
Although the predicted text was not perfect, after adding the preprocessing of the image,
it improved the accuracy of the TrOCR model highly.
